```python
#import csv from file
df = pd.read_csv(r'LT/Hospitals_2018.csv')

print(df)
```

```python
print(df.columns)
```

```python
del df['Subordination: 1-national (MoH), 3-municipality, 8-private, 9-other ministries (not MoH)']
```

```python
#add a column 'country' and fill with 'Lithuania'
df["country"]="Lithuania"
```

```python
#Renaming columns
df.rename(columns = {'ID':'id', 
                     'Name':'name',
                     'Address':'address',
                     'Number of beds at the end of the 2018':'cap_beds',
                     'type_name':'facility_type'},
          inplace = True) 
```

```python
df
```

```python
#create bing variable with API key (insert your own)
bing=Bing(timeout=100,api_key='key')
```

```python
#make an address column which combines name, address, and country
df["Address"]=df["name"]+", "+df["address"]+", "+df["country"]
df
```

```python
#apply the Bing geocoder to 'Address' column and save it in new column 'Coordinates'
df["Coordinates"] = df["Address"].apply(bing.geocode)
```

```python
#extract the lat from 'Coordinates'
df["lat"]=df["Coordinates"].apply(lambda x: x.latitude if x != None else None)
```

```python
#extract the lon from 'Coordinates'
df["lon"]=df["Coordinates"].apply(lambda x: x.longitude if x != None else None)
```

```python
#Save csv to folder
df.to_csv(r'LT/LT_geolocated.csv')
```
